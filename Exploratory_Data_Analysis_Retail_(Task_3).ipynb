**Data Visualization**

```python
# visualizing the dataset as a whole using the pair plot
sb.pairplot(Datafile)
```

```python
# finding the pairwise correlations between the columns and visualising using heatmaps
sb.heatmap(Datafile.corr(), yticklabels=True, cbar=True, cmap='viridis', annot= True)
```

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=90)
ax=sb.countplot(x="Sub-Category", data=Datafile)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
ax=sb.countplot(x="Ship Mode", data=Datafile)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
ax=sb.countplot(x="Segment", data=Datafile)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
ax=sb.countplot(x="Category", data=Datafile)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```

```python
from matplotlib.gridspec import GridSpec
targetCounts = Datafile['Category'].value_counts()
targetLabels  = Datafile['Category'].unique()

plt.figure(1, figsize=(25,20))
the_grid = GridSpec(2, 2)


cmap = plt.get_cmap('coolwarm')
colors = [cmap(i) for i in np.linspace(0, 1, 3)]
plt.subplot(the_grid[0, 1], aspect=1, title='CATEGORY DISTRIBUTION')

source_pie = plt.pie(targetCounts, labels=targetLabels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

```python
from matplotlib.gridspec import GridSpec
targetCounts = Datafile['Region'].value_counts()
targetLabels  = Datafile['Region'].unique()

plt.figure(1, figsize=(25,20))
the_grid = GridSpec(2, 2)


cmap = plt.get_cmap('Paired')
colors = [cmap(i) for i in np.linspace(0, 1, 2)]
plt.subplot(the_grid[0, 1], aspect=1, title='Region')

source_pie = plt.pie(targetCounts, labels=targetLabels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

```python
from matplotlib.gridspec import GridSpec
targetCounts = Datafile['State'].value_counts()
targetLabels  = Datafile['State'].unique()

plt.figure(1, figsize=(35,35))
the_grid = GridSpec(2, 2)


cmap = plt.get_cmap('Paired')
colors = [cmap(i) for i in np.linspace(0, 1, 5)]
plt.subplot(the_grid[0, 1], aspect=1, title='State')

source_pie = plt.pie(targetCounts, labels=targetLabels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

```python
from matplotlib.gridspec import GridSpec
targetCounts = Datafile['Sub-Category'].value_counts()
targetLabels  = Datafile['Sub-Category'].unique()

plt.figure(1, figsize=(25,25))
the_grid = GridSpec(2, 2)


cmap = plt.get_cmap('coolwarm')
colors = [cmap(i) for i in np.linspace(0, 1, 5)]
plt.subplot(the_grid[0, 1], aspect=1, title='Sub-Category')

source_pie = plt.pie(targetCounts, labels=targetLabels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.show()
```

```python
plt.figure(figsize= (30,40))
Datafile.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
ax=sb.countplot(x="Quantity", data=Datafile, hue= 'Region')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
ax=sb.countplot(x="Sub-Category", data=Datafile, hue= 'Region')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```

```python
# computing top sub-categories in terms of sales from first 100 observations
top_subcategory_sales = Datafile.groupby("Sub-Category").Sales.sum().nlargest(n = 100)
# computing top sub-categories in terms of profit from first 100 observations
top_subcategory_profit = Datafile.groupby("Sub-Category").Profit.sum().nlargest(n = 100)

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_sales.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_subcategory_profit.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Sub-Category Sales vs Profit",fontsize = 15)
plt.show()
```
